```python
def check_for_location(x,y,n):
    global grid
    # check row-wise, 
    for col in range(9):
        if grid[x][col] == n:
            return False
    
    # check col-wise
    for row in range(9):
        if grid[row][y] == n:
            return False
    
    # check in box:
    for row in range((x//3)*3,(x//3)*3+3):
        for col in range((y//3)*3,(y//3)*3+3):
            if grid[row][col] ==n:
                return False
    return True
    
```

```python
def solve():
    global grid
    # print("There are {} zeros in grid".format(grid.size - np.count_nonzero(grid)))
    for x in range(9):
        for y in range(9):
            if grid[x][y]==0:
                for n in range(1,10):
                    if check_for_location(x,y,n):
                        grid[x][y]=n
                        # print(grid)
                        #input("Continue?")
                        solve()
                        grid[x][y]=0
                return
    print("Output Grid Solution:")
    print(grid)
    input("More?")
```
